```python
# Fibonacci Recursive

import time

def fibonacci_recursive(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else: 
        return (fibonacci_recursive(n-1) + fibonacci_recursive(n-2)) % 65536

start_time = time.time()
i = 41
result = fibonacci_recursive(i)

print("Compute the " + str(i) + "th Fibonacci number need %d seconds." % (time.time() - start_time))
```

Computed the 41th Fibonacci number need 76 seconds. This is the slowest method to compute Fibonacci number. Its run time is exponential and under one minute only computed the 41th Fiboinacci number.

```python
# Fibonacci Iterative
import time
def fibonacci_iterative(n):
    a = [0,1]
    for i in range(2,n+1):
        f_i = a[i-1] + a[i-2]
        a.append(f_i)
    return a[-1] % 65536    

start_time = time.time()
F_list = [0, 1]
i = 2
while time.time() - start_time < 60:
    F_list.append((F_list[i-1] + F_list[i-2]) % 65536)
    i += 1
    
print("The iterative method gets to the %d th Fibonacci number." % (i))
print("The last Fibonacci number it got to has the mod :" + str(F_list[-1]))
```

The iterative algorithm is achieved by storing previous Fibonacci number in memory. It is $O(n)$ so a lot faster than the recursive method.
